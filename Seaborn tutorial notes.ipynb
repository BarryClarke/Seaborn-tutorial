### Visualising statistical relationships

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

#### The scatter plot

```python
# The scatterplot() is the most basic visualisation and is the default
# plot when relplot() is used
# The tips dataset is in-built in Seaborn, so it understands "tips"
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
# In the above example the total bill vs the tips is plotted. But, in 
# Seaborn, this can be further enhanced by plotting these according to
# a third variable, using a "hue semantic"
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
# We have 3 kind of Semantics 
# 1. Hue - shown above
# 2. Style
# 3. Size
# You can further enhance this by introducing the "style semantic"
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

```python
# You can also represent four variables by changing the hue and style of
# each point independently
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
```

```python
# In the above the hue semantic was categorical, so the default qualitative 
# palette was applied. If the hue semantic is numeric, the default switches
# to a sequential palette
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```python
# You can customise the palette using the string interface cubehelix_palette()
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.7,l=.5", data=tips);
```

```python
# The 3rd kind of semantic is size
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

```python
# You can adjust the sizes by using the "sizes" label
sns.relplot(x="total_bill", y="tip", size="size", sizes=(50, 200), data=tips);
```

#### Line plots

```python
# Earlier, we said scatter plot was the default in replot(). If you want to
# use a line plot, use kind="line"
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```
